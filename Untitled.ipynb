We drop the columns that are meaningless for our study

```python
recipes=data.drop(['URL', 'UserId'], axis=1)
```

What about the units?

```python
recipes['SugarScale'].unique()
```

I rather prefer using Specific Gravity scale, so I will take those units as the unique with the aid of the equation: 

SG = 1 + (plato / (258.6 – ( (plato/258.2) *227.1) ) )

```python
mask = (recipes['SugarScale']=='Plato')
plato_OG=recipes.loc[mask, 'OG']
plato_FG=recipes.loc[mask, 'FG']
plato_BG=recipes.loc[mask, 'BoilGravity']
```

```python
recipes.loc[mask,'OG'] =round(1 + (plato_OG/(258.6-((plato_OG/258.2)*227.1))),3)
recipes.loc[mask,'FG'] =round( 1 + (plato_FG/(258.6-((plato_FG/258.2)*227.1))),3)
recipes.loc[mask,'BoilGravity'] =round( 1 + (plato_BG/(258.6-((plato_BG/258.2)*227.1))),3)
```

Now that we have the same unit for every recipe, we drop that column for making the data frame more readable

```python
recipes=recipes.drop('SugarScale',axis=1)
```

```python
recipes
```

Now that we have our cleaner dataset, let's have a look to the most efficient method (BrewMethod) according to the records by focusing in the average efficiency across the brew methods.
This should be unrelated to the beer style since it is a purely fundamental property, this is, it depends directly to the equipment, the degree of the grain crush, time of mashing process and other variables that are not explicitly represented in the dataset. 

```python
recipes['BrewMethod'].unique()
```

```python
plt.figure(figsize=(12.0,8.0))
recipes.groupby('BrewMethod')['Efficiency'].mean().plot(kind='bar',rot=0)
plt.title('Average efficiency for each of the brew methods', fontsize=20)
plt.ylabel('Efficiency')
plt.show()
```

All Grain and BIAB methods have nearly the same average efficiency. This is not strange since the latter is a particular case of the former, in a nutshell both are an all grain mash. 
Surprisingly, partial mash and extract methods have a really bad overall efficiency, this is rather strange since malt extract should have the enough sugar to increase the specific gravity in the proper amount.
Let's have a look to what is happening. 

It is not a rule but a general convention that a good effiency should be between 70% and 80%, but in our case we take a more flexible range and look for the data where the efficiency is lesser and higher than 65% respectively.

```python
lower_eff = recipes[recipes['Efficiency']<65.0].groupby('BrewMethod').count()
total_count = recipes.groupby('BrewMethod').count()
```

```python
f, ax = plt.subplots(figsize=(12, 8))
sns.set_color_codes("pastel")
sns.barplot(x=total_count['Efficiency'], y=total_count.index, data=total_count, label="Total", color="b")

sns.set_color_codes("muted")
sns.barplot(x=lower_eff['Efficiency'], y=lower_eff.index, data=lower_eff, label="Efficiency<65%", color="b")

ax.set(title='Efficiency distribution per brew method', ylabel='Brew method', xlabel='Count of Beer Recipes')
ax.legend(ncol=1, loc="lower right", frameon=False)
plt.show()
```

Thus even when the beer recipes where extract method was employed are much lesser than those with all grain-like method, the relationship between the recipes that use extract method and reached less than 65% for the efficiency, and the total of the extract-like recipes leads to the fact that extract method is the worst method for brewing beer in terms of efficiency.

By construction, the efficiency is the ratio between the actual sugar extraction that the equipment and method allow to have, and the potential (theoretical) yield that the grain bill could give.

But why is it necessary to consider the efficiency when the brewer is putting all the fermentable sugar with a prefabricated extract after all?
Looking for some randomly selected recipes in the web page I realized that the legend "steeping grains only" is broadly used in these recipes, clarifying that specialty malts are used by steeping them in the wort made with the extract.

Therefore, as a matter of efficiency, extract method leads to a waste of product from specialty malts in the sense that there is left sugar in the grains. This of course does not mean that extract beers are bad, but in terms of harnessing the whole "grain bill" and perhaps saving costs, brewers could take into account to brew with an all grain method.

To finish this efficiency study, let's have a look to the most frequent efficiency obtained with the extract method, and the mean Boil Gravity and Boil Size (sugar concentration and volume of the wort to be boiled) averaged across efficiency values.

```python
#First we select the recipes where extract method was employed,
#take the average of the pre-boil variables per beer style
mask=recipes['BrewMethod']=='extract'
ext_rec=recipes.loc[mask, ['Style', 'BoilSize', 'BoilGravity', 'Efficiency']].groupby('Efficiency').mean()

#We add a column that counts how many recipes were made with malt extract and the percentage they represent
ext_rec['TotalRecipes'] = recipes.loc[mask, ['Style', 'BoilSize', 'BoilGravity', 'Efficiency']].groupby('Efficiency').count()['BoilSize']
ext_rec['TotalRecipes%'] = ext_rec['TotalRecipes']/len(recipes.loc[mask])*100

#Now we take the most frequent efficiencies obtained when extract method was employed
sort_ext_rec = ext_rec.reset_index().sort_values('TotalRecipes%', ascending=False)
ext_rec.sort_values('TotalRecipes%', ascending=False)[:5]
```

```python
f=plt.figure(figsize=(15,5))

#Create the grid
axes1=f.add_subplot(1,3,1)
axes2=f.add_subplot(1,3,2)
axes3=f.add_subplot(1,3,3)

#Fill with plots
sns.barplot(ax=axes1, data=sort_ext_rec[:5], y='Efficiency', x='TotalRecipes%', orient='h', color='b')
sns.scatterplot(ax=axes2, data=ext_rec, x='TotalRecipes', y='BoilSize', color='b')
sns.scatterplot(ax=axes3, data=ext_rec, y='BoilGravity', x='TotalRecipes', color='b')

#Set axis labels
axes1.set_xlabel('Percentage of Recipes')
axes1.set_ylabel( 'Efficiency' )
axes2.set_xlabel('Number of Recipes')
axes2.set_ylabel( 'Boil Size' )
axes3.set_xlabel('Number of Recipes')
axes3.set_ylabel( 'Boil Gravity' )

#Set range of visualization
axes2.set_xlim(0,300)
axes3.set_xlim(0,300)
axes3.set_ylim(1,1.2)

f.tight_layout()

plt.show()
```

In the first figure, we can observe that 35% of efficiency is the most frequent value reported by brewers, in fact 6720 recipes report a 35% of efficiency and an overall Boil Gravity and Boil Size of 1.084 and ~15 Liters respectively. This is quite interesting because it does mean that extract method is broadly used for high sugar density recipes when brewers have small batch sizes, surely this is because brewers' equipment is not big enough to hold the amount of grain that a high gravity recipe needs.
